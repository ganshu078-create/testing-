## 🟢 Setup: Create Table & Insert Data

```python

-- Create Table
CREATE TABLE students (
    student_id INT,
    name VARCHAR(50),
    course VARCHAR(50),
    score INT,
    attendance INT,
    join_date DATE,
    city VARCHAR(50)
);

-- Insert Data
INSERT INTO students VALUES
(1, 'Amit', 'SQL', 85, 90, '2023-06-15', 'Mumbai'),
(2, 'Neha', 'Python', 92, 85, '2023-07-10', 'Delhi'),
(3, 'Rahul', 'SQL', 78, 88, '2023-05-20', 'Mumbai'),
(4, 'Sneha', 'Python', 88, 92, '2023-08-01', 'Bangalore'),
(5, 'Karan', 'SQL', 95, 80, '2023-04-18', 'Mumbai'),
(6, 'Pooja', 'Python', 82, 87, '2023-09-05', 'Delhi');
```

## 1. Ranking students based on score

```python
SELECT name, score,
RANK() OVER (ORDER BY score DESC) AS rank_position
FROM students;
```

## 2. Previous student's score

```python
SELECT name, score,
LAG(score) OVER (ORDER BY score DESC) AS previous_score
FROM students;
```

## 3. Uppercase names & join month

```python
SELECT UPPER(name) AS name_upper,
MONTHNAME(join_date) AS join_month
FROM students;
```

## 4. Next student's attendance

```python
SELECT name, attendance,
LEAD(attendance) OVER (ORDER BY attendance DESC) AS next_attendance
FROM students;
```

## 5. Divide into 4 performance groups

```python
SELECT name, score,
NTILE(4) OVER (ORDER BY score DESC) AS performance_group
FROM students;
```

## 6. Row number per course

```python
SELECT name, course, attendance,
ROW_NUMBER() OVER (PARTITION BY course ORDER BY attendance DESC) AS row_num
FROM students;
```

## 7. Days enrolled

```python
SELECT name,
DATEDIFF('2025-01-01', join_date) AS days_enrolled
FROM students;
```

## 8. Format join date

```python
SELECT name,
DATE_FORMAT(join_date, '%M %Y') AS formatted_date
FROM students;
```

## 9. Replace city

```python
SELECT name,
REPLACE(city, 'Mumbai', 'MUM') AS updated_city
FROM students;
```

## 10. Highest score per course

```python
SELECT name, course, score,
FIRST_VALUE(score) OVER (PARTITION BY course ORDER BY score DESC) AS highest_score
FROM students;
```
